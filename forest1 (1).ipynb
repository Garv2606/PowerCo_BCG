```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('forest.csv')
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0','id'])
df.head()
```

```python
de = pd.get_dummies(df, drop_first=True)
de.head()
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df)
ds = pd.DataFrame(scaled_features, columns=df.columns)
```

```python
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
poly_features = poly.fit_transform(ds)
poly_df = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(ds.columns))
poly_df.head()
```

```python
ds['consumption_sum'] = ds[['cons_12m', 'cons_gas_12m']].sum(axis=1)
ds['consumption_mean'] = ds[['cons_12m', 'cons_gas_12m']].mean(axis=1)
```

```python
ds['consumption_bin'] = pd.cut(ds['cons_12m'], bins = 5, labels = False)
```

```python
x = ds.drop(columns=['churn'])
y = ds['churn']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 42, test_size = 0.25)
```

```python
y_train_ = pd.cut(y_train, bins=2, labels=[0, 1])
y_test_ = pd.cut(y_test, bins=2, labels=[0, 1])
```

```python
model = RandomForestClassifier(random_state = 42)
```

```python
model.fit(x_train, y_train_)
```

```python
sel = SelectFromModel(model, prefit = True)
```

```python
x_train_sel = sel.transform(x_train)
x_test_sel = sel.transform(x_test)
```

```python
model.fit(x_train_sel, y_train_)
```

```python
y_pred = model.predict(x_test_sel)
y_pred
```

```python
model.score(x_test_sel, y_test_)
```

```python
from sklearn import metrics
```

```python
tn, fp, fn, tp = metrics.confusion_matrix(y_test_, y_pred).ravel()
```

```python
print(tn, fp, fn, tp)
```

```python
print(metrics.accuracy_score(y_test_, y_pred))
print(metrics.precision_score(y_test_, y_pred))
print(metrics.recall_score(y_test_, y_pred))
```
